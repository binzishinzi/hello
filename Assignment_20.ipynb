```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the data
iris_df = pd.read_csv("Iris.csv")
```

```python
# Display the first few rows of the dataset
print(iris_df.head())
```

```python
# List down the features and their types
feature_types = iris_df.dtypes
print("\nFeatures and their types:")
print(feature_types)
```

```python
# Create histograms for each feature
for column in iris_df.columns[1:5]:  # Excluding the 'Id' column and 'Species' column
    plt.figure(figsize=(8, 6))
    sns.histplot(iris_df[column], kde=True)
    plt.title(f'Histogram of {column}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Create boxplots for each feature
for column in iris_df.columns[1:5]:  # Excluding the 'Id' column and 'Species' column
    plt.figure(figsize=(8, 6))
    sns.boxplot(y=iris_df[column])
    plt.title(f'Boxplot of {column}')
    plt.ylabel('Value')
    plt.show()
```
