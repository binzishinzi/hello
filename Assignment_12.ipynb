Scan the Iris dataset and give the inference as:
1. List down the features and their types (e.g., numeric, nominal) available in the dataset.
2. Create a histogram for each feature in the dataset to illustrate the feature distributions.
3. Create a boxplot for each feature in the dataset.
4. Compare distributions and identify outliers.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv("Iris.csv")
```

```python
# List down the features and their types
print(df.info())
```

```python
# Create histograms for each feature
numeric_features = df.select_dtypes(include=['float64']).columns
for feature in numeric_features:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[feature], kde=True)
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Create boxplots for each feature
for feature in numeric_features:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x=feature, data=df)
    plt.title(f'Boxplot of {feature}')
    plt.xlabel(feature)
    plt.show()
```

To compare distributions and identify outliers, you can visually inspect the boxplots generated for each numeric feature. Outliers are data points that fall significantly outside the range of the majority of the data points on the boxplots.

Here's how you can interpret the boxplots to identify outliers:

Outliers below the lower whisker: These are data points that fall below the lower whisker of the boxplot. They are usually represented as points outside the lower fence (Q1 - 1.5 * IQR), where Q1 is the first quartile and IQR is the interquartile range. These are considered low outliers.
Outliers above the upper whisker: These are data points that fall above the upper whisker of the boxplot. They are usually represented as points outside the upper fence (Q3 + 1.5 * IQR), where Q3 is the third quartile and IQR is the interquartile range. These are considered high outliers.
