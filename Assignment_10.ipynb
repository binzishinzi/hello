```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic_data = pd.read_csv('titanic.csv')
```

```python
# Set the style for seaborn
sns.set_style("whitegrid")
```

```python
# Plot survival counts by gender and passenger class
plt.figure(figsize=(12, 6))
sns.countplot(data=titanic_data, x='Pclass', hue='Survived', palette='coolwarm')
plt.title('Survival Counts by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right')
plt.show()
```

```python
# Plot survival counts by gender
plt.figure(figsize=(8, 6))
sns.countplot(data=titanic_data, x='Sex', hue='Survived', palette='Set1')
plt.title('Survival Counts by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right')
plt.show()
```

```python
# Plot survival counts by age
plt.figure(figsize=(12, 6))
sns.histplot(data=titanic_data, x='Age', hue='Survived', multiple='stack', kde=True, palette='muted')
plt.title('Survival Counts by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right')
plt.show()
```

```python
# Plot histogram of ticket prices
plt.figure(figsize=(10, 6))
sns.histplot(data=titanic_data, x='Fare', bins=30, kde=True, color='skyblue')
plt.title('Distribution of Ticket Prices')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

Certainly! Histograms are a type of graphical representation that display the distribution of numerical data. They are particularly useful for visualizing the frequency or density distribution of continuous variables, such as the distribution of ticket prices in the Titanic dataset.

Here's a breakdown of the theory behind plotting a histogram:

Data Binning: The first step in creating a histogram is to divide the range of values into intervals, or "bins." Each bin represents a range of values, and the data points falling within that range contribute to the count or frequency of that bin.

Frequency Count: Once the data is divided into bins, the next step is to count the number of data points that fall into each bin. This count represents the frequency of occurrence of values within that particular range.

Visualization: The frequency counts are then plotted as bars on a graph, where the x-axis represents the value ranges (bins), and the y-axis represents the frequency or count of data points falling within each bin. The height of each bar indicates the frequency of data points in that bin.

Visualization Options: Histograms can also include additional visual elements, such as a line or curve representing the density estimation (KDE - Kernel Density Estimate), which provides a smoothed estimate of the underlying distribution of the data.

In the context of the Titanic dataset:

The 'Fare' column represents the ticket prices paid by passengers.
By plotting a histogram of the 'Fare' column, we can visualize the distribution of ticket prices across all passengers.

The histogram allows us to see how the ticket prices are spread out, which can provide insights into the range, central tendency, and variability of ticket prices among passengers on the Titanic.

Overall, histograms are a valuable tool for exploring and understanding the distribution of numerical data, making them widely used in data analysis and visualization tasks.
